```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
```

```python
#import training set
mydata = pd.read_csv("F:/DataSet/Iris_data.csv")
mydata
```

```python
Y=mydata.Species
Y
```

```python
X=mydata.iloc[:,1:5]
X

```

```python
mymodel = MultinomialNB()
mymodel.fit(X,Y)
```

```python
accuracy=mymodel.score(X,Y)
accuracy
```

```python
test =pd.read_csv("F:/DataSet/Iris_test.csv")
test
```

```python
test_x = test.iloc[:, 1:5]
test_y = test.Species
```

```python
test_accuracy = mymodel.score(test_x,test_y)
round(test_accuracy*100,3)
```

```python
predtest = mymodel.predict(test_x)
predtest
```

```python
from sklearn.metrics import confusion_matrix
mat=confusion_matrix(test_y, predtest)
mat
```

```python
from sklearn.metrics import classification_report
report = classification_report(test_y,predtest)
print(report)
```
